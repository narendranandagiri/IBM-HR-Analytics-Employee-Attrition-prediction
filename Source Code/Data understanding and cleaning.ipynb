#                                                       Data Cleaning

By this we can determine some of the columns that are not important for analysing and modelling.

They are:

1. EmployeeCount: This feature has only one unique value which is not important.
2. EmployeeNumber: This feature is just a unique number of the employees.
3. Over18: since all the employees are over18 there is no need of this feature.
4. StandardHours: for evey employee the standard hours are 80.

These columns are to be removed.


There are some columns initially numerical but they are actually categorical, They should be replaced by a suitable names.

They are:

1. Education
2. EnvironmentSatisfaction
3. JobInvolvement
4. JobLevel
5. JobSatisfaction
6. PerformanceRating
7. RelationshipSatisfaction
8. WorkLifeBalance
9. StockOptionLevel

For the StockOptionLevel describes High-level employees of a company, such as executives or managers who are typically given the option to purchase shares of the company at a set price, which is usually lower than the market price. So it basically gives the level of StockOption. It may come under categorical column which is of Level 0, Level 1, Level 2, Level 3.

```python
df = df.drop(columns = ['EmployeeCount','EmployeeNumber','Over18','StandardHours'], axis = 1)
```

```python
df['Education'] = df['Education'].apply(lambda x : 'Schooling' if x == 1 else ( 'Intermediate' if x == 2 else ( 'Bachelors' if x == 3 else ('Masters' if x == 4 else 'PhD'))))
```

```python
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].apply(lambda x : 'Low' if x == 1 else ( 'Medium' if x == 2 else ( 'High' if x == 3 else 'Very high')))
```

```python
df['JobInvolvement'] = df['JobInvolvement'].apply(lambda x : 'Low' if x == 1 else ( 'Medium' if x == 2 else ( 'High' if x == 3 else 'Very high')))
```

```python
df['JobLevel'] = df['JobLevel'].apply(lambda x : 'Fresher' if x == 1 else ( 'Junior' if x == 2 else ( 'Senior' if x == 3 else ('Team Lead' if x == 4 else 'Manager'))))
```

```python
df['JobSatisfaction'] = df['JobSatisfaction'].apply(lambda x : 'Low' if x == 1 else ( 'Medium' if x == 2 else ( 'High' if x == 3 else 'Very high')))
```

```python
df['PerformanceRating'] = df['PerformanceRating'].apply(lambda x : 'Poor' if x == 1 else ( 'Good' if x == 2 else ( 'Excellent' if x == 3 else 'Outstanding')))
```

```python
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].apply(lambda x : 'Poor' if x == 1 else ( 'Good' if x == 2 else ( 'Excellent' if x == 3 else 'Outstanding')))
```

```python
df['WorkLifeBalance'] = df['WorkLifeBalance'].apply(lambda x : 'Poor' if x == 1 else ( 'Good' if x == 2 else ( 'Better' if x == 3 else 'Best')))
```

```python
df['StockOptionLevel'] = df['StockOptionLevel'].apply(lambda x : 'Level 0' if x == 0 else ( 'Level 1' if x == 1 else ( 'Level 2' if x == 3 else 'Level 3')))
```
